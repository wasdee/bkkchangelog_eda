```python
import pandas as pd

df = pd.read_json('Datasets/backup.ndjson.tar.gz', lines=True)
```

```python
# make all columns title case
df.columns = df.columns.str.title()
```

```python
df.Count_Reopen.unique()
```

```python
df.dtypes
```

```python
# turn Cords into tuples
df["Coords"] = df.Coords.apply(lambda x: tuple(x))
```

```python

# turn Problem_Type_Abdul into sorted tuples
df["Problem_Type_Abdul"] = df.Problem_Type_Abdul.apply(lambda x: tuple(sorted(x)))
```

```python
df.head()
```

```python
# count of unique values in each column
df.nunique()
```

```python
# turn State into a category
df.State = df.State.astype('category')
```

```python
# turn Star to a category
df.fillna(value={'Star': '-1'}, inplace=True)
df.Star = df.Star.astype('int').astype('category')
```

```python
# convert 'last_activity' to datetime
df.Last_Activity = pd.to_datetime(df.Last_Activity)

# convert type org to category
df.Org = df.Org.astype('category')

# convert count_reopen to int
df.fillna({'Count_Reopen': 0}, inplace=True)
df.Count_Reopen = df.Count_Reopen.astype('int')

# convert star to int
df.Star = df.Star.astype('float64')
```

```python
df.to_parquet('Datasets/backup.parquet')
```
